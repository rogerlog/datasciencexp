```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dados_mercado = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mercado.csv', sep=';')
```

```python
dados_mercado
```

```python
len(dados_mercado)
```

```python
print(dados_mercado.values[1,1])
```

```python
lista_compras = []
for i in range(len(dados_mercado)):
  lista_compras.append([str(dados_mercado.values[i, j]) for j in range(4)])
```

```python
lista_compras
```

```python
te = TransactionEncoder()
te_array = te.fit(lista_compras).transform(lista_compras)
te_array
```

```python
df_prod_mercado = pd.DataFrame(te_array, columns = te.columns_)
df_prod_mercado
```

```python
df_prod_mercado.drop(columns = 'nan', inplace = True)
```

```python
df_prod_mercado
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(df_prod_mercado, min_support = 0.3, use_colnames = True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(df_prod_mercado,
                            min_support = 0.3,
                            use_colnames = True)
```

```python
frequent_itemsets
```

## Regras de Associação

```python
from mlxtend.frequent_patterns import association_rules
```

```python
regras = association_rules(frequent_itemsets,
                           metric = 'confidence',
                           min_threshold=0.8)
regras
```

```python
regras = regras[['antecedents','consequents','support','confidence','lift']]
```

```python
regras = regras.rename(columns={"antecedents": "se", 
                                "consequents": "então",
                                "support": "suporte",
                                "confidence": "confiança",
                                "lift": "LIFT"})
```

```python
regras
```

```python
regras.sort_values(by = 'LIFT', ascending = False)
```
